```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_inline import backend_inline
from mpl_toolkits.mplot3d import Axes3D  
#
def plot_3d(function, x_range=(-5, 5, 100), y_range=(-5, 5, 100), xlabel="X", ylabel="Y", zlabel="Z"):
    
    # 
    x = np.linspace(x_range[0], x_range[1], x_range[2])
    y = np.linspace(y_range[0], y_range[1], y_range[2])
    X, Y = np.meshgrid(x, y)
    
    Z = function(X, Y)
    
    # 3D
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # 
    surf = ax.plot_surface(X, Y, Z, edgecolor='none', alpha=0.8)
    
    # 
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_zlabel(zlabel)
    
    plt.show()

# 
def f(x, y):
    return 3*x**3+ y**4+ np.exp(y)+5

# 
plot_3d(f, x_range=(-5, 5, 100), y_range=(-5, 5, 100))
```

```python
import numpy as np
import matplotlib.pyplot as plt

#
np.random.seed(42)  
n_points = 1000
x = np.random.uniform(-2, 2, n_points)
y = np.random.uniform(-2, 2, n_points)
Z = 3*x**3 + y**4 + np.exp(y) + 5 + np.random.normal(0,0.01) 
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
    

# 
scatter = ax.scatter(x, y, Z, marker='o', alpha=0.8)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
 
    
plt.show()

#
def model(x, y, a, b, c, d):
    return a*x**3 + b*y**4 + c*np.exp(y) + d
#
a= np.random.normal (2.5,0.01,1)
b= np.random.normal (2.5,0.01,1)
c= np.random.normal (2.5,0.01,1)
d= np.random.normal (2.5,0.01,1)

losses =[]
```

```python
lr = 0.02
num_epochs = 200
k =0.001

#
for epoch in range(num_epochs):
    Z_hat = model(x,y,a,b,c,d)
    #
    loss =  np.mean((Z_hat-Z)**2)
    losses.append(loss)
    #
    e = Z_hat - Z
    da = np.mean(2 * e * x**3)
    db = np.mean(2 * e * y**4)
    dc = np.mean(2 * e * np.exp(y))
    dd = np.mean(2 * e)
    #
    #a = a- lr * da
    a = (1-k*lr)*a- lr * da
    b = (1-k*lr)*b- lr * db
    c = (1-k*lr)*c- lr * dc
    d = (1-k*lr)*d- lr * dd
    #20
    if epoch % 20 == 0:
        print(f'epoch {epoch }, loss: {loss:.5f}')
```

```python
#
print(f"a = {a.item():.6f} (true: 3.0)")
print(f"b = {b.item():.6f} (true: 1.0)")
print(f"c = {c.item():.6f} (true: 1.0)")
print(f"d = {d.item():.6f} (true: 5.0) ")
```

```python
#
plt.figure(figsize=(10, 6))
plt.plot(losses)
#xy
plt.xlabel('epoch', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```
